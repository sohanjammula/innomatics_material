### What is RAG (Retrieval Augmented Generation)?

**RAG** is a mechanism that enhances the quality of Large Language Models (LLMs) by connecting
them to external knowledge sources in real-time.
- Before generating a response, the system first retrieves relevant, up-to-date information and
provides it to the LLM as context, which then augments the model's ability to generate a more
accurate, detailed, and trustworthy answer.

**RAG** is a powerful technique that enhances large language models (LLMs) by **combining them with external knowledge sources** during text generation.

**The Problem with Traditional LLMs:**
- Rely **solely on internal training data**
- Knowledge is **limited and quickly outdated**
- Prone to **hallucinations** and factual errors



**How RAG Solves This:**
```
Query → Retrieve relevant docs → Augment LLM prompt → Accurate response
```

**Core Process:**
1. **Retrieve**: Pull relevant information from external sources (databases, documents, web)
2. **Augment**: Combine retrieved data with user prompt  
3. **Generate**: LLM produces **grounded, up-to-date answers**




#### What is a Vector Database?

A vector database is a specialized database designed to store, index, and retrieve data in the form of vector embeddings—numerical representations that capture the semantic meaning of data such as text, images, or audio.


**Key Aspects of Vector Databases**

1. Semantic Search

  Vector databases enable semantic search, allowing queries based on meaning and context rather than exact keyword matches. This makes them highly effective for applications like question answering, recommendation systems, and retrieval-augmented generation (RAG).

2. Built for AI-Native Workloads

  These databases are inherently optimized for AI-driven applications, where data is processed and understood in embedding space. They seamlessly integrate with machine learning models and support workflows that rely on contextual understanding.

3. Scalability and Speed

  Vector databases are designed to handle large-scale, high-dimensional data efficiently. They use advanced indexing techniques to ensure fast similarity search, even across millions or billions of vectors.

Some widely used vector databases include:
	•	Pinecone
	•	Chroma
	•	Redis (with vector extensions)
	•	Elasticsearch / OpenSearch

Additionally, traditional databases and data warehouses such as PostgreSQL, Oracle, BigQuery, and Snowflake have introduced support for vector embeddings, enabling hybrid use cases.

**Key Advantages:**
**Real-time knowledge** - Access current data beyond training cutoff
**Domain accuracy** - Works with enterprise documents and proprietary data  
**Reduced hallucinations** - Answers based on actual retrieved content
**No retraining needed** - Cost-effective solution

RAG transforms static LLMs into **dynamic knowledge systems** by retrieving relevant external information during generation, making it ideal for Q&A, search, and enterprise applications requiring fresh, accurate data.

### Why use RAG?

- **Limited Knowledge:** LLMs are trained on a fixed dataset, making them susceptible to outdated or incomplete information.
- **Hallucinations:** LLMs can generate plausible-sounding but incorrect or nonsensical information, especially when faced with ambiguous queries or unfamiliar topics.
- **Lack of Contextual Awareness:** LLMs may struggle to understand the specific context of a query, leading to irrelevant or inaccurate responses.

**RAG addresses these limitations by providing access to up-to-date and contextually relevant information, improving the accuracy and reliability of LLM-generated outputs.**

**What are the benefits of RAG?**

- **Improved Accuracy and Relevance:** RAG enables LLMs to generate more accurate and relevant responses by leveraging external knowledge sources.
- **Reduced Hallucinations:** By grounding responses in retrieved information, RAG helps minimize the generation of incorrect or nonsensical information.
- **Enhanced Contextual Understanding:** RAG allows LLMs to better understand the specific context of a query, leading to more meaningful and informative responses.
- **Adaptability:** RAG can be easily adapted to various domains and tasks by plugging in different retrieval systems and knowledge sources.

**What are the appplications of RAG?**

- **Question Answering:** Provide accurate and comprehensive answers to questions by retrieving relevant information from vast knowledge bases.
- **Chatbots and Virtual Assistants:** Enhance the capabilities of chatbots and virtual assistants by providing them with access to up-to-date and contextually relevant information.
- **Content Generation:** Generate high-quality and informative content by leveraging external sources for research and fact-checking.
- **Summarization:** Create concise and accurate summaries of long documents or articles by extracting key information from the text.

### Function Calling

**Function calling in LLMs enables models to interact with external tools and APIs.** It allows the LLM to analyze a user prompt, decide when a function is needed, and generate the precise parameters required to invoke it effectively.

**Core Mechanism**
The process starts by providing the LLM with descriptions of available functions, including their names, input arguments (with data types), and expected outputs (also with data types). The model then parses the user's query to determine if any function should be called and how to structure the call. This makes LLMs more capable of handling real-world tasks beyond pure text generation.

**Common External Tools**
- APIs and HTTP endpoints for web services.
- Databases queried via Python functions.
- Search engines accessed through SDKs or libraries.
- Any custom logic wrapped in functions.

**Why It Matters**

Function calling forms a foundation for agentic frameworks like AWS Bedrock or Google's agent kits, powering orchestration in LLM-driven apps by bridging natural language with structured actions.



### Orchestration

Orchestration is the process of coordinating various components to execute complex tasks.

**Orchestration coordinates multiple components—like function calls—to handle complex tasks in LLM applications.** It determines the sequence or parallelism of actions, ensuring the right tools are invoked at the right time for accurate results.

**Why Orchestration Matters?**

- **Not all data suits vectorization**, such as transactional records in MySQL or reporting from BigQuery/Snowflake, so orchestration lets LLMs access structured sources via functions without dumping everything into vectors.

- **It extends LLM capabilities** by connecting to diverse tools (APIs, databases) in sequence while preserving context.
- **This boosts accuracy** by minimizing hallucinations and supports intricate workflows, like checking order status *and* inventory stock across separate functions.


**User asks:**
`What's my last order status, and is that item back in stock?`

Orchestration invokes `fetch_user_orders` for status and `fetch_inventory` for availability—potentially in parallel—then synthesizes responses, unlike single-function limits.

### What is agentic AI?

**Agentic AI refers to AI systems that autonomously make decisions and take actions to achieve goals with minimal human intervention.** These systems use AI agents—software that leverages LLMs to perform tasks on behalf of users, going beyond simple automation.

**Key Characteristics**
- **Autonomy**: Operate independently, perceiving environments and acting without constant oversight.
- **Goal-driven**: Pursue specific objectives through planning and execution.
- **Multi-step reasoning**: Break down complex problems into sequences of actions.
- **Adaptability**: Learn from interactions, context, and past experiences to improve.

#### Why Agentic AI Excels Over Basic Setups?

- Simple RAG or function calling works for single data sources, like product lookups in a vector store or basic database queries.

- Agentic AI shines with scattered data (NoSQL, vectors, documents, transactions) and multi-turn conversations—it deploys specialized agents with defined tools, handles orchestration automatically, and maintains context without manual routing.

**Real-World Use Cases**
- Customer service: Resolve inquiries across channels, schedule meetings, monitor workflows.
- E-commerce: Product recommendations, post-sales follow-ups, inventory checks.

#### Part 4: Create a front end using Streamlit

##### Use AWS Bedrock for Retreival and Generation

```python
import os
import streamlit as st
import boto3 # pip install boto3
from dotenv import load_dotenv

load_dotenv()

st.set_page_config(page_title="Simple KB Chat", page_icon="💬")
st.title("Simple Bedrock Knowledge Base Chat")

# Check your AWS region
region = os.getenv("AWS_REGION", "eu-north-1")
# Check your knowledgebase ID
knowledge_base_id = 'DDSGHZ5O1L'
# Check your model ARN # amazon.nova-2-lite-v1:0
model_arn = "eu.amazon.nova-2-lite-v1:0"

st.write("Ask a question to your Amazon Bedrock Knowledge Base.")

question = st.text_input("Enter your question")

if st.button("Ask"):
    if not knowledge_base_id:
        st.error("Set KNOWLEDGE_BASE_ID as an environment variable.")
    elif not model_arn:
        st.error("Set MODEL_ARN as an environment variable.")
    elif not question.strip():
        st.error("Please enter a question.")
    else:
        client = boto3.client("bedrock-agent-runtime", region_name=region)

        response = client.retrieve_and_generate(
            input={"text": question},
            retrieveAndGenerateConfiguration={
                "type": "KNOWLEDGE_BASE",
                "knowledgeBaseConfiguration": {
                    "knowledgeBaseId": knowledge_base_id,
                    "modelArn": model_arn,
                },
            },
        )

        answer = response["output"]["text"]
        st.subheader("Answer")
        st.write(answer)
```

#### Use Amazon bedrock Knowledge base for retreival and Generate using Gemini

```python
import os
import boto3
import streamlit as st
from google import genai
from dotenv import load_dotenv

load_dotenv()

st.set_page_config(page_title="KB + Gemini Chat", page_icon="💬")
st.title("Bedrock KB + Gemini 2.5 Flash")

# Check your AWS region
aws_region = os.getenv("AWS_REGION", "eu-north-1")
# Check your knowledge base ID
knowledge_base_id = 'DDSGHZ5O1L'
os.environ['GEMINI_API_KEY'] = os.getenv("gemini_key")

st.write("This app retrieves context from Amazon Bedrock Knowledge Base and uses Gemini 2.5 Flash to answer.")

question = st.text_input("Enter your question")

if st.button("Ask"):
    if not knowledge_base_id:
        st.error("Set KNOWLEDGE_BASE_ID as an environment variable.")
    elif not question.strip():
        st.error("Please enter a question.")
    else:
        bedrock = boto3.client("bedrock-agent-runtime", region_name=aws_region)

        kb_response = bedrock.retrieve(
            knowledgeBaseId=knowledge_base_id,
            retrievalQuery={"text": question},
            retrievalConfiguration={
                "vectorSearchConfiguration": {
                    "numberOfResults": 5
                }
            },
        )

        chunks = []
        for item in kb_response.get("retrievalResults", []):
            text = item.get("content", {}).get("text", "")
            if text:
                chunks.append(text)

        context = "\n\n".join(chunks)

        prompt = f"""
Answer the user's question using only the context below.

Context:
{context}

Question:
{question}
"""

        client = genai.Client()
        response = client.models.generate_content(
            model="gemini-2.5-flash",
            contents=prompt
        )

        st.subheader("Answer")
        st.write(response.text)

        with st.expander("Retrieved Context"):
            st.write(context if context else "No context returned.")
```
