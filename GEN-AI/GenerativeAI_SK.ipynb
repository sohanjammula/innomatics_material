# Generative AI

## Introduction to Generative AI

### What is Generative AI?

Artificial intelligence systems that can produce high quality content, specifically **text, images, and audio**.

**Generative AI (GenAI)** enables machines to autonomously create original content—mimicking human creativity and cognition. Modern GenAI systems produce human-quality **images, essays, poems, music, and videos** on demand.

GenAI exploded into **public awareness** with OpenAI's ChatGPT launch in November 2022. Users could ask open-ended questions like *"Write a limerick about an office party"* or *"How do you reverse an array in Python?"* and receive coherent, context-aware responses.

**ChatGPT's growth was unprecedented**: It hit **100 million users in just 2 months**—faster than TikTok (9 months, leveraging China's Douyin) or Instagram (2.5 years). This marked the moment GenAI transitioned from research labs to mainstream adoption.

Generative Al is also a developer tool

Generative AI is inherently stochastic: the same prompt can yield different outputs each time, and minor phrasing tweaks often produce varied results. This unpredictability drives prompt engineering—the skill of carefully crafting inputs to consistently elicit optimal responses.

### Why Generative AI?

**Generative AI's rapid adoption** stems from two key factors: **breakthrough technical capabilities** and **universal accessibility**. Recent advances have made GenAI powerful enough to transform everyday workflows.

**Zero barriers to entry**: Anyone—technical experts or complete beginners—can generate professional-quality **text, images, or music** with simple clicks, democratizing creativity like never before.

**Industry transformation potential**:
- **Customer Support**: Personalized, 24/7 chatbot experiences
- **Marketing**: Instant high-quality campaigns and copy
- **Entertainment**: Realistic VFX, character animation, synthetic actors

GenAI doesn't just replace tasks—it **supercharges human productivity** across every sector.

- **Job displacement concerns**: Rapid automation of **creative, writing, design, and analytical tasks** threatens entire job categories from copywriting to graphic design

**Dual-edged sword**: Same technology enabling personalized customer support can create undetectable propaganda. Same tools revolutionizing marketing can eliminate marketing jobs.

### How Generative AI different from ML (DL)?

Machine learning models generally fall into three categories:
1. discriminative,
2. reinforcement, and
3. generative.

Discriminative models classify or distinguish between different types of data instances, making them ideal for predicting labels or values on new data points.
Reinforcement learning, by contrast, skips classification altogether instead, it relies on a reward signal as feedback. The model iteratively learns through trial and error to maximize cumulative long-term rewards across a sequence of actions.

Generative models, meanwhile, learn the probability distribution from existing data to create entirely new instances, enabling them to produce novel content.

#### Large Language models (What? and Why?)

Large language models (LLMs) are advanced NLP AI systems, built on transformer architectures with billions of parameters, trained on massive text datasets to process and generate natural language. They use attention mechanisms to capture complex language relationships by predicting the next word in sequences.

A single LLM handles diverse NLP tasks effectively, even with little or no task-specific data, delivering strong results from simple prompts.

- process and generate natural language text
- trained on massive amounts of text data
- identify complex relationships in language
- transformer model

Unlike traditional models like RNNs—which don't scale well, require huge task-specific labeled data (e.g., for sentiment analysis, translation), need separate models per task (question answering, summarization, etc.), and fail at multi-step reasoning—LLMs excel across diverse NLP tasks with a single model, even with minimal or no domain-specific training data, delivering strong results from simple prompts.

- limited scale and performance
- enormous training examples needed
- separate models for different NLP tasks
- not cater to multi-step reasoning tasks

### How do large language models (LLMs) generate text, and what role does supervised learning play?

LLMs generate text by completing prompts like
- "My favourite food is" → next word "roti";
- "My favourite food is roti" → next word "with";
- "My favourite food is roti with" → next word "dal."

They use supervised learning to predict the next word repeatedly, trained on massive datasets (hundreds of billions to trillions of words) from internet text.

A single sentence like "My favourite food is roti with dal" becomes multiple training examples to learn these patterns. This enables LLMs application like ChatGPT to produce coherent continuations, useful for writing, information lookup, and brainstorming.

**Make a Completion Request**

Create a completion request to the OpenAI API using the `chat.completions.create` method.

To generate a response to a prompt, we use the `client.chat.completions.create` method.

`client.chat.completions.create` needs at minimum two arguments. The `model` to be used to generate your response and the `messages` to respond to.

This method takes two keyword arguments.
1. The model keyword argument is mandatory and
2. The prompt is specified using the messages keyword argument.

Typically, a conversation is formatted with `a system message first`, followed by `alternating user and assistant messages`.

Each message in the list should have two pieces of information:
- who said it (either the “system,” the “user,” or the "assistant") and
  - The `system role` is typically used for directing the model; it defines the model's purpose, persona, output format, and any other necessary instructions.
  - The `user role`, it is you. The content you pass as a user can range from a question you need answered to a sentence you want extended, to a paragraph requiring grammar corrections, and much more.
  - Another type of message we can include in our prompt to the model is an `assistant message`. Assistant messages can be used whenever we want to provide the model with example outputs to give additional context or guide it toward the writing style.
- what they said (the actual words they used).

````python
# Legacy/Standard
response = client.chat.completions.create(
    model = 'gpt-5-nano',
    messages = [
        {"role": "system", "content": "You are a teacher, really great at explaining programming concepts in a few words."},
        {"role": "user", "content": "Explain what is an API"}
        ])
response
````

We find that **completion** is an instance of the **ChatCompletion** class, with the parameter **choices** storing a list of **Choice** objects that, in turn, keep the responses from the chatbot. By default, the model returns only one response, which is typically the desired setting. This number can be controlled through a parameter called **n**. Of course, more responses result in higher token consumption and, therefore, higher cost.

````python
print(response.choices[0].message.content)
````

````python
# Streaming output
response = client.chat.completions.create(
    model = 'gpt-5-nano',
    messages = [
        {"role": "system", "content": "You are a teacher, really great at explaining programming concepts in a few words."},
        {"role": "user", "content": "Explain what is an API"}
        ],
    stream = True)
for chunk in response:
  print(chunk.choices[0].delta.content, end = ' ')
````

### Temperature

**How temperature impacts the generated response.**

Temperature defaults to 1 and accepts values between  and  inclusive. Try changing the value of temperature from the default of 1.

````python
from openai import OpenAI

client = OpenAI()

def get_response(prompt, temperature):
    completion = client.chat.completions.create(
        model="gpt-4o-mini",
        messages=[
            {"role": "user", "content": prompt}
        ],
        temperature=temperature
    )
    return completion.choices[0].message.content
````

When **temperature is set to 0** it is referred to as `argmax sampling`, meaning the option with the highest probability is always selected.

- The option with the highest probability will perceive as the most “correct” answer.
- Higher temperatures will generate a more diverse response.

Temperature controls how much randomness is in the output.
- Lower temperature means less randomness which is suitable for contexts which require more stable output.
- Higher temperature means more randomness which is suitable for more creative applications.

````python
prompt = "Write a tagline for an ice cream shop"
response = get_response(prompt, 0)
print(response)
````

### Tokens

The GPT family of models process text using tokens. Tokens are numerical representations of words or characters.

- The process of tokenization involves dividing plain text, such as a phrase, sentence, paragraph, or entire documents into smaller chunks of data so that it is easier to analyze.
- These smaller chunks of data are referred to as tokens, which may include words, phrases, or sentences.
- In English, 1 token is approximately equal to 4 characters

Token usage is tracked in several categories:
- **Input tokens** – tokens in your request.
- **Output tokens** – tokens generated in the response.
- **Cached tokens** – reused tokens in conversation history (often billed at a reduced rate).
- **Reasoning tokens** – in some advanced models, extra “thinking steps” are included internally before producing the final output.

#### Tokenizing by word

The benefit of using tokenization by word is that we can pinpoint words that are frequently used. If we were analyzing a group of restaurant ads in NY, and found that the word `vegan` was used often, then we might assume that there are plenty of vegan options at these restaurants.

- Here are some helpful rules of thumb for understanding tokens in terms of lengths:
  - 1 token ~= 4 chars in English
  - 1 token ~= ¾ words
  - 100 tokens ~= 75 words

#### Tokenizing by sentence

When tokenizing by sentence, we have the ability to analyze how the words in the sentence correlate with each other, which allows us to better understand the context of the words. If we were analyzing a group of restaurant ads in NY, and found that the sentence `No vegan options`. was used, then we can determine that there are not plenty of vegan options at these restaurants.

- Here are some helpful rules of thumb for understanding tokens in terms of lengths:
  - 1-2 sentence ~= 30 tokens
  - 1 paragraph ~= 100 tokens
  - 1,500 words ~= 2048 tokens

#### `max_tokens` parameter

GPT takes the prompt, converts the input into a list of tokens, processes the prompt, and converts the predicted tokens back to the words we see in the response.

- Keep track of the following when using the API:
  - **Completions** - depending on the engine used, requests can use up to  tokens shared between prompt and completion.
  - **For specialized endpoints** - Answers, Search, and Classifications - the query and longest document must be below  tokens together.

### Top P

`top_p`, an alternative to sampling with temperature, is also referred to as nucleus sampling.
- Generally, it is not recommended to alter both the temperature and the top_p.
- `top_p` controls how many random results should be considered for completion as per the temperature.
- If we set top_p to 0.1, it means only the tokens comprising the top 10% probability mass are considered.

````python
from openai import OpenAI

client = OpenAI()

prompt = "Write a tagline for an ice cream shop"
responses = client.chat.completions.create(
    model="gpt-4o-mini",
    messages=[
        {"role": "user", "content": prompt}
    ],
    n=5,
    top_p=0.1
).choices

for response in responses:
    print(response.message.content.strip())
````

- We can use our probabilities tool to see the probability associated with each token.

- Higher temperatures will generate a more diverse response. However top_p will narrow the scope of possible answers to the top 10%. Randomly choosing between the top 10% will give us more or less the same completion.

- Here we can see that the response generated 5 times was the same. That is because it is taking the top 10 most probable responses which is the same. What happens when we switch `top_p = 0.8`.

````python
responses = client.chat.completions.create(
    model="gpt-4o-mini",
    messages=[
        {"role": "user", "content": prompt}
    ],
    n=5,
    top_p=0.8
).choices

for response in responses:
    print(response.message.content.strip())
````

**In short, the top_p parameter controls the diversity of the generated responses, with lower values leading to more repetitive responses and higher values leading to more varied responses.**

Think of frequency_penalty as a way to not have too many same-word repetitions. ## Customizing LLMs

Transform generic API responses into exactly what you need – without building models from zero.
- Pretrained large language models like GPT-4o, Gemini 3 Flash, or Llama 3.2 are incredibly capable but often produce generic, verbose, or irrelevant responses when used with basic prompts.
- Customization transforms these general-purpose models into specialized tools tailored to your exact needs.

Here are the **four proven methods** to customize pretrained models without training from scratch.
1. Controlling Generation Parameters (Instant Precision)
2. Prompt Engineering (The 90% Solution)
3. Prompt Chaining (Complex Task Mastery)
4. Fine-Tuning (Domain Specialization)

### Controlling Generation Parameters

Fine-tunes the model's **creativity vs determinism** using hyperparameters.

**Key Parameters:**

- **Temperature (0-2):** Controls randomness
  - 0.1: Precise, repetitive (exam answers, code generation)
  - 0.7: Balanced creativity (lesson planning)
  - 1.5+: Highly creative (brainstorming, poetry)
- **Top-P (0-1):** Nucleus sampling – probability mass cutoff
- **Max Tokens:** Strict length control

### Prompt Engineering

Crafts precise instructions to guide the model exactly where you want it to go.

**5 Control Dimensions**
- **FORMAT:** "List 5 points" vs "Write 300-word essay"
- **LENGTH:** "One sentence" vs "Detailed explanation"  
- **STYLE:** "Formal letter" vs "Casual explanation" vs "JSON output"
- **TONE:** "Professor to student" vs "Friend explaining to friend"
- **CONTEXT:** "CBSE Class 10" vs "IIT-JEE aspirant"

### Prompt Chaining

Breaks complex problems into sequential simple steps. Thereby, single prompts overwhelm models on multi-part tasks.
- Through prompt chaining, complex tasks are broken down into sub-tasks.
- Each sub task has its own `prompt + last output`
- Each prompt build upon previous outputs, can incrementally solve a task
- Final output is a result of cumulative work

**Example:**
1. "Suggest 3 Optuna hyperparameters for XGBoost"
2. "Write MLflow logging code for those params"
3. "Add validation metrics plotting"
4. "Create complete training pipeline"

Trains model on YOUR data to create domain expert.

## Controlling Generation Parameters

The Foundational language models are next-token predictors, which means that,
given the previously generated text, they **return a probability distribution for the next token to be generated.**
- While generating text, LLMs perform next word prediction based on previous generated text.

LLMs generate text by predicting the next token in a sequence, completing prompts step by step like this:

1. "My favorite food is" → next token "pizza";
2. "My favorite food is pizza" → next token "and";
3. "My favorite food is pizza and" → next token "soda."

- Naive approach to predict next word is, choose the most likely word based on the probability.
- Always picking max probability creates boring, repetitive text. Sophisticated/rare words get ignored. Only toy models use this `greedy` approach.

**What happens if we select the next word randomly from the possible words, rather than always picking the most likely one?**

In pure sampling, more probable words are still chosen more often, but less likely words get a chance too. Even if "pizza" is selected first (as it's the top choice), the next word might be "calculations." This makes the generated text more diverse and interesting.

However, pure sampling has a downside: the text often gets into unusual or niche topics. While the chance of picking a rarely used word at any one step is low, the probability of *some* rarely used word appearing anywhere in the sequence is much higher. For example, if esoteric words collectively account for 20% probability at each step $ p = 0.2 $, the chance that *none* appear in a 5-word sequence is $ (1 - 0.2)^5 = 0.8^5 \approx 0.33 $. Thus, the probability of at least one rarely used word is $ 1 - 0.33 = 67\% $.

**Next Token Selection**

The model assigns probabilities to possible next tokens based on training data and selects one, usually with techniques like:
1. **Greedy Search:** Always picking the most likely token.
2. **Beam Search:** Exploring multiple token sequences before selecting the best one.
3. **Temperature Scaling:** Adjusting randomness in token selection to control creativity.
4. **Top-p Sampling:**
3. **Top-k Sampling:** Limiting choices to the top k most probable tokens.

### [Temperature](https://shivammehta25.github.io/posts/temperature-in-language-models-open-ai-whisper-probabilistic-machine-learning/)

One effective way to address rare word continuations in sampling is by applying **temperature**.

- Temperature, **a scaling factor** that adjusts the probability distribution. In our example, while some food item dominate as top continuations, a long tail of rarer words collectively holds substantial probability mass.

- **Lowering temperature** (e.g., to near 0 or 0.1) sharpens the distribution: it **suppresses low-probability (rare words) tails—since probabilities must sum to 1—boosting top choices even more.** Use low temperature for deterministic, training-data-faithful output like code generation or factual responses.

- **Raising temperature** (e.g., to 0.8 or 1.5+) flattens the distribution: **it downweights high-probability peaks while elevating rarer options, promoting creativity.** Opt for higher values in brainstorming, essays, or poetry where diverse training data recombination shines.

Temperature thus balances **coherence and novelty**, favoring familiar patterns at low values or exploration at high ones.

### Top K Sampling

**What is the problem with temperature?**

Often, you'll want creativity beyond the top 1-2 most likely words, without drifting into bizarre combinations. Temperature alone falls short here—a low value suppresses rare words but doesn't eliminate them entirely.

Top-K vs. Temperature
| Method | Limits Choices By | Strength | Limitation |
|--------|-------------------|----------|------------|
| Low Temperature | Sharpens full distribution | Predictable focus | Rare words persist |
| Top-K (e.g., k=4) | Hard cutoff at top k | Blocks tail outliers | Fails if >k good options

**Top-K sampling** addresses this by capping consideration to the *k* most probable options at each step.

- For k=4, the model samples proportionally only from the top 4 continuations, effectively renormalizing their probabilities.

This shines when distributions are peaked with few strong options, acting as a guardrail against oddities in high-stakes scenarios like code.

````python
# Top-K = 3 (limits to top 3 most likely words)
for _ in range(3):
    response = client.models.generate_content(
        model='gemini-2.5-flash-lite-preview-09-2025',
        contents='Write a single tagline for an ice cream shop',
        config=genai.types.GenerateContentConfig(
            temperature=1,  # Moderate creativity
            top_k=3           # Sample ONLY from top 3 continuations
        )
    )
    print(response.text)
````

### Top-p (Nucleus) Sampling

**What is the problem with top-K sampling?**

It adapts better than top-k (avoids under/over-inclusion) and pairs with temperature: **first rescale probabilities, then apply the cutoff** for balanced creativity without extremes.

One way to address this is to set the threshold not on the number of words, but on their cumulative probability. This is formally called nucleus sampling, or informally, top-P sampling.

- If you set top_p to 0.8, you are asking the GenAI tool to choose the **smallest list of words that together cover 80% of the probabilities**, and discard the rest.

**Top-p (Nucleus) Sampling** offers a flexible alternative to top-k by filtering words based on cumulative probability rather than a fixed count, using "We are going to __" as an example where many mid-tier options (walk, talk, hang) compete equally.

For top_p=0.8, the model includes the smallest set of highest-probability words summing to 80% mass, discarding the tail—dynamic across contexts: many words for flat distributions, few for peaked ones.

````python
# Top-p = 0.9 (uses smallest set of words summing to 90% probability)
for _ in range(3):
    response = client.models.generate_content(
        model='gemini-2.0-flash',
        contents='Write a single tagline for an ice cream shop',
        config=genai.types.GenerateContentConfig(
            temperature=1,  # Base creativity
            top_p=0.9         # Dynamic cutoff at 90% mass
        )
    )
    print(response.text)
````

#### Self-Consistency Chain-of-Thought

- Performs multiple CoT experiments (Experiments should be diverse)
- Improves coherence and Consistency
- Use voting algorithm to pick most consistent answer - applies idea of ensemble learning to LLMs

**Example:**
*“When Michael was ten years old, his brother was two-thirds his age. If Michael is now 40, how old is his brother?”*

`The prompt encourages the model to “think step by step`

If we run this prompt several times, we may get **different answers**. For example, one run might output **37**, another might output **35.5**, and a third run might return **37** again. Because the answers vary, we apply a **self-consistency approach**: instead of relying on a single chain of thought, we sample multiple reasoning paths and then use **majority voting** to determine which final answer appears most often.

In this case, **37** is generated in two out of the three runs, so the voting mechanism selects **37** as the final answer.


````python
user_prompt = 'When Michael was ten years old, his brother was two-thirds his age. If Michael is now 40, how old is his brother?'

def chain_of_thought(user_prompt):
  response = client.chat.completions.create(
      model = 'gpt-5-nano',
      messages = [
          {'role': 'system', 'content': "You are a helpful assistent and answer precise and concise. Return ages in nearby integers in years"},
          {'role': 'user', 'content': f"{user_prompt}, think step-by-step."}
      ]

  )
  response_text = response.choices[0].message.content
  return response_text
````

````python
print(chain_of_thought(user_prompt))
````

````python
print(chain_of_thought(user_prompt))
````

````python
# Self Consistency CoT
def self_consistency_chain_of_thought(user_prompt, number_of_chains=3):
    # run CoT multiple times
    res = []
    for i in range(number_of_chains):
        res_chain = chain_of_thought(user_prompt)
        res.append(res_chain)
        print(f"{i}: {res_chain}")

    # Ask the model to return most common answer
    res_concat = ";".join(res)
    self_consistency_prompt = (
    f"You will receive several model answers inside << >>, separated by ';'. "
    f"<<{res_concat}>> "
    "Each answer contains step-by-step reasoning and a final stated answer. "
    "Your task:\n"
    "1. Ignore all reasoning steps.\n"
    "2. Extract ONLY the final explicit answer from each response (usually after 'Answer:').\n"
    "3. Determine which final answer appears most frequently.\n"
    "4. Return ONLY that answer, exactly as originally written, with no extra commentary."
    )
    response = client.chat.completions.create(
        model = 'gpt-5-nano',
        messages = [
            {"role": "user",
             "content": f"{self_consistency_prompt}"}
            ]
        )
    response_text = response.choices[0].message.content
    return f'Final Response is: {response_text}'
````

````python
user_prompt = 'When Michael was ten years old, his brother was two-thirds his age. If Michael is now 40, how old is his brother?'
print(self_consistency_chain_of_thought(user_prompt, number_of_chains=3))
````

#### [Tree of Thoughts](https://arxiv.org/pdf/2305.10601)

- When humans face complex decisions, we naturally explore multiple options, evaluate risks, and adjust our choices when new information appears. **Tree of Thought (ToT) Prompting** brings this same flexible decision-making process to AI systems.

- A framework that generalizes over chain-of-thought prompting and encourages exploration over thoughts `(unlike CoT, model do not stick to single idea suitable for stright forward tasks, it explores multiple reasoning paths)` that serve as intermediate steps for general problem solving with language models.

**What is essential for ToT-style reasoning?**

1. Encouraging exploration of possibilities,
2. Listing explanations, and
3. Weighing pros and cons prompts the model to use branching logic. Allows backtracking when a particular path is flawed.

image.png

* **Root:** Begin with a single prompt or problem.
* **Branches:** Generate several initial ideas
  *Example: Reducing traffic in a city → better public transit, tolls, carpooling*.
* **Sub-branches:** Each idea develops deeper reasoning

  * Public transit → cost, timeline, commuter behavior
  * Tolls → congestion pricing, fairness
* **Pruning:** Weak or irrelevant branches are discarded.
* **Final Selection:** The most balanced, evidence-based answer is chosen.

It resembles constructing and evaluating a decision tree.

**How to Prompt Tree of Thought Reasoning?**

To encourage ToT behavior:

1. **Ask the model to explore multiple possibilities**

   * “List several explanations,” not just one.
2. **Evaluate each idea**

   * Require pros and cons for every path.
3. **Encourage comparison and selection**

   * Ask the model to choose the best option with justification.
4. **Allow pruning of weaker ideas**

   * The model can discard unhelpful or inconsistent paths.

This promotes critical thinking and structured decision-making.

**Game of 24**

The *Game of 24* is a classic mathematical reasoning task in which the objective is to combine four given numbers using basic arithmetic operations (+, −, ×, ÷) to produce the value **24**. For instance, given the input **“4 9 10 13”**, one valid solution is:

$$
(10 − 4) × (13 − 9) = 24

#### Self Feedback

- [SELF-REFINE: Iterative Refinement with Self-Feedback](https://arxiv.org/pdf/2303.17651)
- This technique iteratively improves quality of response

````python
import json
import re
````

````python
system_prompt = '''
You will receive an input inside << >>. Your task is to evaluate how well the input explains the major events and historical importance of the Indian Independence movement. While evaluating, consider:

1. How broad and detailed the historical context is
2. Whether the major turning points and key figures are covered
3. How well short-term and long-term consequences are explained
4. Clarity, completeness, and accuracy

After evaluating, if you notice missing points, shallow explanations, or unclear reasoning, return a JSON output in the following format:

{
  "rating": "score in percentage",
  "feedback": "specific and detailed feedback describing what needs improvement",
  "revised_output": "an improved and more complete explanation of the key events and significance of the American Civil War. Avoid using apostrophes (') and double quotes (\")."
}

'''
user_prompt = "The Revolt of 1857 was a fight by Indian soldiers against the British rule."
revised_output = "The response is sufficient."
feedback = ''
target_rating = 90
max_iterations = 3

for i in range(max_iterations):
  response = client.chat.completions.create(
      model = 'gpt-5-nano',
      messages = [
          {"role": "system", "content": system_prompt},
          {"role": "user", "content": user_prompt if i == 0 else f"Feedback: {feedback}, Revised Output: {revised_output}"}
          ])
  response_text = response.choices[0].message.content

  # convert to valid json
  valid_json = response_text
  response_json = json.loads(valid_json)

  # extract rating
  rating_str = response_json['rating']
  rating_num = int(re.findall(r'\d+', rating_str)[0])

  # extract feedback
  feedback = response_json['feedback']

  # extract revised output
  revised_output = response_json['revised_output']
  print(f"i={i}, Rating: {rating_num}, \nFeedback: {feedback}, \nRevised Output: {revised_output}")
  if (rating_num >= target_rating) or (i == max_iterations-1):
      print('\nFinal Output', revised_output)
````

#### [Self Critique](https://cobusgreyling.medium.com/self-critique-llm-chain-using-langchain-smartllmchain-d67c42a4fa83)

This technique follows a three step procedure
- [1] **ideation step** gets a predefined number of output proposals (ideas) from the LLM.

- A LLM then [2] **critiques** all of the ideas looking for possible flaws in the proposals and picking the most appropriate suggestion.

- Lastly, in [3] **resolve**, the LLM tries to improve the best idea from the [2] *critique* step. The output here constitutes the final answer.

### ReAct Prompting

A framework, where LLMs are used to generate both **reasoning traces** and **task-specific actions** in an interleaved manner.

The ReAct framework can allow LLMs to **interact with external tools** to retrieve additional information that leads to more reliable and factual responses.

**How it Works?**

ReAct is inspired by the synergies between "acting" and "reasoning" which allow humans to learn new tasks and make decisions or reasoning.

Chain-of-thought (CoT) prompting has shown the capabilities of LLMs to carry out reasoning traces to generate answers to questions involving arithmetic and commonsense reasoning, among other tasks. But its lack of access to the external world or inability to update its knowledge can lead to issues like fact hallucination and error propagation.

You're a software engineer on a mission: build an AI agent that doesn't just respond—it thinks. In this lab, you'll step into the role of an AI architect, designing a smart assistant that solves tough problems by reasoning through them and taking purposeful actions.

Using the ReAct (Reasoning + Acting) framework, you'll teach your agent to think step by step, consult tools like search engines or calculators, and adapt on the fly. It’s not just about answers—it’s about how the agent gets there.

By the end of the lab, your AI will face a mystery that can’t be solved with knowledge alone. It will need logic, resourcefulness, and the ability to act—just like you, the engineer who built it.

**What is ReAct?**

**ReAct** stands for **Reasoning + Acting**. It's a framework that combines:

1. **Reasoning**: The agent thinks through problems step by step, maintaining an internal dialogue about what it needs to do.
2. **Acting**: The agent can use external tools (search engines, calculators, APIs) to gather information or perform actions.
3. **Observing**: The agent processes the results from its actions and incorporates them into its reasoning.

This creates a powerful loop: **Think → Act → Observe → Think → Act → ...**

**Why ReAct Matters**

Traditional language models are limited by their training data cutoff and can't access real-time information. ReAct agents overcome this by:
- Accessing current information through web searches
- Performing calculations with specialized tools
- Breaking down complex problems into manageable steps
- Adapting their approach based on intermediate results

**What is Agent State?**

In ReAct, state management is crucial, as the agent must maintain context across multiple reasoning and acting steps.

````python
from typing import (Annotated,Sequence,TypedDict)
from langchain_core.messages import BaseMessage
from langgraph.graph.message import add_messages

class AgentState(TypedDict):
  """The state of the agent."""

  # add_messages is a reducer
  messages: Annotated[Sequence[BaseMessage], add_messages]
````

**Key Concepts:**
- **State**: Contains the conversation history and context.
- **Reducer**: `add_messages` automatically handles adding new messages to the conversation.
- **Type Safety**: TypedDict ensures our state structure is well-defined.

````python
# Example conversation flow:
state: AgentState = {"messages": []}

# append a message using the reducer properly
state["messages"] = add_messages(state["messages"], [HumanMessage(content="Hi")])
print("After greeting:", state["messages"])

# add another message (e.g. a question)
state["messages"] = add_messages(state["messages"], [HumanMessage(content="Who is Sir C.V Raman?")])
print("After question:", state)
````

1. The user asks a complex question requiring current data.
2. The model analyzes the query and realizes it needs to search for information.
3. The model generates a tool call for the search.

````python
#  Initial Query Processing
dummy_state: AgentState = {
    "messages": [HumanMessage("Who won ICC Women world cup in 2025?")]}

response = react_model.invoke({"scratch_pad":dummy_state["messages"]})

dummy_state["messages"]=add_messages(dummy_state["messages"],[response])
response
````

1. Extract the tool call from the model's response.
2. Execute the tool using the specified arguments.
3. Create a ToolMessage containing the results.
4. Add the tool result to the conversation state.

````python
# Tool Execution
tool_call = response.tool_calls[-1]
print("Tool call:", tool_call)

tool_result = tools_by_name[tool_call["name"]].invoke(tool_call["args"])
print("Tool result preview:", tool_result)

tool_message = ToolMessage(
    content=json.dumps(tool_result),
    name=tool_call["name"],
    tool_call_id=tool_call["id"]
)
dummy_state["messages"] = add_messages(dummy_state["messages"], [tool_message])
````

````python
# Processing Results and Next Action
response = react_model.invoke({"scratch_pad": dummy_state["messages"]})
dummy_state['messages'] = add_messages(dummy_state['messages'], [response])

# check if the model wants to use another tool
if response.tool_calls:
    tool_call = response.tool_calls[0]
    tool_result = tools_by_name[tool_call["name"]].invoke(tool_call["args"])
    tool_message = ToolMessage(
        content=json.dumps(tool_result),
        name=tool_call["name"],
        tool_call_id=tool_call["id"]
    )
    dummy_state['messages'] = add_messages(dummy_state['messages'], [tool_message])
````

````python
# Final Response Generation
response = react_model.invoke({"scratch_pad": dummy_state["messages"]})
print("Final response generated:", response.content is not None)
print("More tools needed:", bool(response.tool_calls))
````

#### ReAct with Graphs

LangGraph automates this process with a state machine that handles the reasoning loop automatically.

- Automatically execute all tool calls from the model
- Handle multiple simultaneous tool calls
- Return properly formatted tool messages

````python
def tool_node(state: AgentState):
    """Execute all tool calls from the last message in the state."""
    outputs = []
    for tool_call in state["messages"][-1].tool_calls:
        tool_result = tools_by_name[tool_call["name"]].invoke(tool_call["args"])
        outputs.append(
            ToolMessage(
                content=json.dumps(tool_result),
                name=tool_call["name"],
                tool_call_id=tool_call["id"],
            )
        )
    return {"messages": outputs}
````

- Call the ReAct-enabled model
- Pass the full conversation context
- Return the model's response (which may include tool calls)

````python
# Model Invocation Node
def call_model(state: AgentState):
    """Invoke the model with the current conversation state."""
    response = react_model.invoke({"scratch_pad": state["messages"]})
    return {"messages": [response]}
````

- Implement the control flow logic
- Decide whether the agent needs to use more tools
- Route the conversation to either tool execution or completion

````python
def should_continue(state: AgentState):
    """Determine whether to continue with tool use or end the conversation."""
    messages = state["messages"]
    last_message = messages[-1]
    # If there is no function call, then we finish
    if not last_message.tool_calls:
        return "end"
    # Otherwise if there is, we continue
    else:
        return "continue"
````

1. **Agent Node**: Where reasoning happens and tool calls are generated.
2. **Tools Node**: Where tool execution occurs.
3. **Conditional Edge**: Determines whether to continue or finish.
4. **Entry Point**: Conversation always starts with the agent reasoning.

````python
from langgraph.graph import StateGraph, END

# Define a new graph
workflow = StateGraph(AgentState)

# Define the two nodes we will cycle between
workflow.add_node("agent", call_model)
workflow.add_node("tools", tool_node)

# Add edges between nodes
workflow.add_edge("tools", "agent")  # After tools, always go back to agent

# Add conditional logic
workflow.add_conditional_edges(
    "agent",
    should_continue,
    {
        "continue": "tools",  # If tools needed, go to tools node
        "end": END,          # If done, end the conversation
    },
)

# Set entry point
workflow.set_entry_point("agent")

# Compile the graph
graph = workflow.compile()
graph
````

````python
def print_stream(stream):
    """Helper function for formatting the stream nicely."""
    for s in stream:
        message = s["messages"][-1]
        if isinstance(message, tuple):
            print(message)
        else:
            message.pretty_print()

inputs = {"messages": [HumanMessage(content="What are the highlights in Indian Financial 2025 budget?")]}

print_stream(graph.stream(inputs, stream_mode="values"))
````
