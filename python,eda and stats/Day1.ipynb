```python
#WAP to mimic the code of capitalize method

-Check if the string is empyty -> if it is empty return the string as it is
-Take the first character -> Convert it into upper case, if the first character is in upper case, leave it as it is
-Take the remaining characters and check if any character is in uppercse -> then convert it into lower case but if it is already in lower case leave it as it is
```

```python
def mimic_capitalize(s):

    # If the string is empty, return it 
    if len(s) == 0:
        return s
    else:
        result = '' # store the output 
        
        # Step-1:Capitalize the first letter
        fc = s[0]

        if 'a'<=fc<='z': # manually checking lowercase 
            result += chr(ord(fc)-32)
        else:
            result += fc

        # Step-2: make all the remaing character lowercase 
        for i in s[1:]:
            if 'A'<=i<='Z': # manually check for the uppercase letter 
                result += chr(ord(i)+32)
            else:
                result += i

    return result
```

```python
def mimic_title(s):
    result = ''
    word = ''
    if len(s) == 0:
        return s
    for i in s:
        if i != ' ':
            word += i
        else:
            if len(word) > 0:
                # Step-1: capitalize first letter
                fc = word[0]
                if 'a' <= fc <= 'z':
                    result += chr(ord(fc) - 32)
                else:
                    result += fc
                # Step-2: lowercase remaining lettes
                for ch in word[1:]:
                    if 'A' <= ch <= 'Z':
                        result += chr(ord(ch) + 32)
                    else:
                        result += ch
                word = ''
            result += ' '
    # Last word
    if len(word) > 0:
        fc = word[0]
        if 'a' <= fc <= 'z':
            result += chr(ord(fc) - 32)
        else:
            result += fc
        for ch in word[1:]:
            if 'A' <= ch <= 'Z':
                result += chr(ord(ch) + 32)
            else:
                result += ch
    return result
```

```python
def mimic_swapcase(s):
    result = ""
    for ch in s:
        # Lowercase to uppercase
        if 'a' <= ch <= 'z':
            result += chr(ord(ch) - 32)
        # Uppercase to lowercase
        elif 'A' <= ch <= 'Z':
            result += chr(ord(ch) + 32)
    return result
```

```python
mimic_title("she sells sea shells on the sea shore")
mimic_swapcase("she sells sea shells on the sea shore")
```
