### Pandas DataFrame
A Pandas DataFrame is a two-dimensional, size-mutable, and potentially heterogeneous tabular data structure with labeled axes (rows and columns). It can be thought of as a collection of Series objects that share the same index

### Features of Pandas DataFrame
- Two-Dimensional : Contains rows and columns
- Heterogenous Data : Each column can contain different types of data
- Labeled Axes : Both rows and columns are labeled
- Flexible Data Manipulation : Supports filtering, grouping and transforming data
- Data Integration : Easily integrates with various data sources and formats
- Handling Missing Data : Functions to detect, fill or drop missing values
- Data Aggregation : Methods for grouping, merging and aggregating data

## Dataframe

- Dataframe is the most widely used data-structure in data analysis. 

- `It is a table with rows and columns, with rows having an index and columns having meaningful names.`

**Syntax to create dataframe**

> df = pd.DataFrame(`list,Series,array,dictionary`)

```python
import pandas as pd

# Sample dictionary where keys are row indices
data = {
    'Row1': ["John", 24, "B.Com"],
    'Row2': ["Nick", 25, "B.E."],
    'Row3': ["Tom", 28, "B.Arch"]
}
# Creating DataFrame with dictionary keys as columns
df = pd.DataFrame.from_dict(data, orient='index', columns=['Name', 'Age', 'Qualification'])

df
```

```python
dict = {
    "Name" : ["John", "Nick", "Tom"],
    "Age" : [24, 25, 28],
    "Qualification" : ["B.Com", "B.E." , "B.Arch"]
}

df = pd.DataFrame(dict, index = ["Row1" , "Row2", "Row3"])
df
```

**Creating a DataFrame using Dictionaries**

```python
dict_1 = {
    "Name" : ["A","B","C","D"],
    "Designation":["PM","Dev","TM","Tester"],
    "Location":["Delhi", "Hyderbad", "Chennai","Vizag"],
    "Salary LPA":[20, 12, 13,9]
}
```

```python
df2 = pd.DataFrame(dict_1)
df2
```

**Creating indexed dataframe**

```python
df3 = pd.DataFrame(dict_1, index = ["person"+f" {i}" for i in range(1,1+len(dict_1["Name"]))])
```

**Creating a DataFrame with NaN values**   

- NaN value------ Null value `syntax for NaN -- np.nan`

```python
import numpy as np
import pandas as pd
```

```python
dict_1 = {
    "Name" : ["A",np.nan,"C","D"],
    "Designation":["PM",np.nan,"TM","Tester"],
    "Location":["Delhi", "Hyderbad", "Chennai","Vizag"],
    "Salary LPA":[20, 12, 13,9]
}
```

```python
df4 = pd.DataFrame(dict_1)
df4
```

```python
df4.to_csv("df4.csv", index = False)
```
