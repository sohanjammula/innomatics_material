## Introduction
**What is Pandas?**

- Pandas is an extremely popular open source library built upon Numpy, for `handling tabular data, data manipulation and analysis.` 

- Probably the best thing about Pandas is that it stores data as a Python object with rows and columns, very similar to data stored in Excel files. Also, this way we can easily visualize our data, making our job a lot easier then handling data in form of lists or dictionaries.

- It is built on the Numpy package and its key data structure is called the **`DataFrame`** also the advantage over` Numpy is that it handles multiple data types (i.e., strings), not only numerical data.

## Here are just a few of the things that pandas does well:
- **DataFrame and Series**  
   Pandas provides two main structures: **DataFrame** (2D labeled, size-mutable table) and **Series** (1D labeled array) to handle heterogeneous data efficiently.

- **Handling Missing Data**  
   Pandas makes handling missing data straightforward with functions to **detect**, **remove**, or **fill** missing values (`NaN`) in both floating point and non-floating point data.

- **Data Alignment**  
   Pandas ensures automatic or explicit **alignment** of data, allowing seamless operations across different datasets. It automatically aligns on labels during computations.

- **Reshaping**  
   Tools for **reshaping** data include pivoting, stacking, unstacking, and transposing. This is essential for reorienting data structures easily.

- **Data Aggregation**  
   Pandas supports **grouping**, **merging**, and **aggregating** data through its powerful `groupby`, `merge`, and aggregation functions. It allows complex transformations in a few lines of code.

- **High Performance**  
   Optimized with **NumPy**, Pandas offers high performance for large datasets, ensuring efficient memory usage and fast computation.

---


### Additional Capabilities of Pandas

- **Easy handling of missing data**  
   Pandas handles missing values (`NaN`) efficiently across different data types, offering methods like `fillna()`, `dropna()`, and `isna()`.

- **Size mutability**  
   DataFrames and other structures in Pandas allow for **insertion** and **deletion** of columns and rows, making them highly adaptable.

- **Automatic and explicit data alignment**  
   When performing operations, Pandas aligns data automatically or based on specific labels, preventing mismatched operations.

- **Group by functionality**  
   Pandas' **groupby()** function is powerful for split-apply-combine operations, enabling data **aggregation** and **transformation** easily.

- **Data conversion**  
   Pandas allows seamless conversion of **NumPy arrays** and other Python structures into **DataFrames** for structured data manipulation.

- **Label-based slicing and indexing**  
   Pandas provides **intelligent label-based slicing** with `.loc[]`, **fancy indexing**, and **subsetting** for working with large datasets.

- **Merging and joining**  
   Intuitive functions like `merge()` and `concat()` make combining datasets, whether by rows or columns, straightforward.

- **Flexible reshaping**  
   Pandas includes **pivoting** and **reshaping** tools like `pivot()`, `melt()`, and `stack()` for dynamic data presentation.

- **Hierarchical labeling**  
   Supports multi-level indexing (MultiIndex) for **hierarchical labeling**, allowing complex data manipulations with ease.

- **Robust I/O tools**  
   Pandas provides built-in methods to **read** and **write** data from/to multiple formats like **CSV**, **Excel**, **SQL databases**, and **HDF5**.

- **Time series functionality**  
   Pandas has specific tools for **time series**, including date range generation, frequency conversion, and **moving window statistics**.

## Series
- A series is similar to a 1-D numpy array, and contains scalar values of the same type (numeric, character, datetime etc.). 
A dataframe is simply a table where each column is a pandas series.

### Features of a Series
- One-dimensional : Contains a single array of data
- Homogeneous data : All elements are of same data type
- Index : Each element in a series associated with an index
- Flexible Indexing : Supports both integer-based and label-based indexing
- Vectorized operations : Supports element-wise opeations and broadcasting
- Handling missing data : Functions to detect, fill or drop missing values
- Statistical operations : Method to compute summary statistics

### Creating Pandas Series

- Series are one-dimensional array-like structures, though unlike numpy arrays, they often contain non-numeric data (characters, dates, time, booleans etc.)

- You can create pandas series from array-like objects using ```pd.Series()```.

- We use the `pd.Series()` command, and provide a list,numpy,dictionary

#### Explicitly specifying indices

You might have noticed that while creating a series, Pandas automatically indexes it from 0 to (n-1), n being the number of rows. But if we want, we can also explicitly set the index ourselves, using the ‘index’ argument while creating the series using `pd.Series()`

````python
### Create an series with indices as name and marks as values
### Use dictionary for creating series
marks = {"Ram":100, "Navi" : 90, "Praveen" :80}
names_marks = pd.Series(marks)
names_marks
````
