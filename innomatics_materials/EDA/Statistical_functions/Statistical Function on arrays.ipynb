## Statistical Functions

NumPy offers a powerful suite of statistical functions that allow for efficient computation of basic statistics on arrays. These functions can be applied to entire arrays or along specific axes for multi-dimensional data.

- np.sum(): Computes the sum of array elements over a given axis.
- np.min() and np.max(): Find the minimum and maximum values in an array.
- np.mean(): Calculates the arithmetic mean along the specified axis.
- np.median(): Computes the median along the specified axis.
- np.var(): Computes the variance of an array.
- np.std(): Computes the standard deviation.
- np.corrcoef(): Returns the correlation matrix, which measures the correlation strengths between variables.

```python
#np.sum(): Computes the sum of array elements over a given axis.
# Creating an array
arr = np.array([[1, 2], [3, 4],[5,6]])

# Sum of all elements
total_sum = np.sum(arr)
print("Total sum:", total_sum)

# Sum along the first axis (sum of each column)
column_sum = np.sum(arr, axis=0)
print("Sum along columns:", column_sum)

# Sum along the second axis (sum of each row)
row_sum = np.sum(arr, axis=1)
print("Sum along rows:", row_sum)
```

```python
# Minimum of all elements
min_val = np.min(arr)
print("Minimum value:", min_val)

# Maximum of all elements
max_val = np.max(arr)
print("Maximum value:", max_val)
```

```python
# Mean of all elements
mean_val = np.mean(arr)
print("Mean value:", mean_val)

# Median of all elements
median_val = np.median(arr)
print("Median value:", median_val)
```

```python
# Variance of all elements
variance = np.var(arr)
print("Variance:", variance)

# Standard deviation of all elements
std_dev = np.std(arr)
print("Standard deviation:", std_dev)
```

```python
# Creating two arrays
x = np.array([1, 2, 3, 4, 5])
y = np.array([5, 4, 3, 2, 1])

# Compute correlation matrix
corr_matrix = np.corrcoef(x, y)
print("Correlation matrix between x and y:\n", corr_matrix)
```
