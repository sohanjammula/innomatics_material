```python
# Bivariate Analysis of Age(continous) and Survival(categorical) data
# Pick a plot that we use for continous data distr and repeat it for every category
# We chose to create KDE plot for Age
# we'll create 2 KDE plots 1. Survival = 0 and 2. Survival = 1
```

```python
df_1 = df.dropna(subset=['Age'])
```

```python
sns.kdeplot(x='Age',data=df,hue='Survived')
```

```python
sns.kdeplot(x='Age',data=df_1,hue='Pclass',palette='Set2')
```

```python
# '''
# Observations
# 1. Central Tendency class 1 ~ 35-40
# 2. Central Tendency class 2 ~ 25-30
# 3. Central Tendency class 3 ~ 25
# '''
```

```python
## compute mean,std dev of Age wrt Pclass
df_1.groupby('Pclass')['Age'].agg(['mean','std'])
```

```python
## Visual Representation
sns.countplot(x='Pclass',data=df)
plt.show()
```
