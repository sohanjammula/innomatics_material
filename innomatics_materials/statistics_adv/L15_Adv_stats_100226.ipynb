**Problem Statement** : A resturant owner asserts that avg total bill for customer is 20. This was an older analysis that he had done, now he wants to validate if the avg is still same or has differred significantly. He'll do hypothesis testing to validate the change. He needs significance level of 0.05. His population standard deviation is 8.5 . Now he takes a random sample, and checks if the avg has changed significantly or its still nearly the same.

```python
'''
Step 1:
As we are validating a change or deviation, we need to do two-tailed test

H1 : The average has changed and its not 20 now. -> state of effect
avg != 20

H0 : The average has not changed and is still 20. -> state of no effect / change
avg = 20


Step 2:
Take a random sample and compute sample statistics (sample_mean,sample_std_dev)

- sample_size = 30
- sample_mean = 22
- sample_std_dev = 9

Step 3:
Compute test statistics (Z/t)
- Here we'll compute Z statistic as population std dev is known

- Compute Z score for given significance level = 0.05

'''
```

```python
import numpy as np
import scipy.stats as stats

pop_mean = 20
pop_std_dev = 8.5
sample_size = 30
sample_mean = 22
sample_std_dev = 9

alpha = 0.05

# Z
z_score = (sample_mean - pop_mean)/(pop_std_dev/np.sqrt(sample_size))

p_value = 2*(1-stats.norm.cdf(z_score))


# Z(alpha/2)
z_critical = stats.norm.ppf(1-alpha/2)

print(f"Z Score : {z_score}")
print(f"Z Critical : {z_critical}")


if np.abs(z_score)>z_critical:
  print("Decision : Accept H1 or Reject H0")
else:
  print("Decision : Failed to reject H0")


if p_value < alpha:
  print("Decision based on p-value : Accept H1 or Reject H0")
else :
  print("Decision based on p-value : Failed to reject H0")
```

Based on the given data and two tailed test, we can say that the change in average is not statistically significant. That is the average bill is still nearby 20.

Problem Statement : A resturant owner asserts that avg total bill for customer is 20. This was an older analysis that he had done, now he wants to validate if the average has increased significantly or not. He'll do hypothesis testing to validate the change. He needs significance level of 0.05. His population standard deviation is 8.5 . Now he takes a random sample, and checks if the avg has increased significantly or not increased.

```python
'''
Step 1:
As we are validating an increase, we need to do one-tailed test, right tailed test

H1 : The average has increased than 20 -> state of effect / change
avg > 20

H0 : The average is less than or equal to 20 -> state of no effect / change
avg <= 20


Step 2:
Take a random sample and compute sample statistics (sample_mean,sample_std_dev)

- sample_size = 30
- sample_mean = 23
- sample_std_dev = 7

Step 3:
Compute test statistics (Z/t)
- Here we'll compute Z statistic as population std dev is known

- Compute Z Critical value for given significance level = 0.05

'''
```

```python
import numpy as np
import scipy.stats as stats

pop_mean = 20
pop_std_dev = 8.5
sample_size = 30
sample_mean = 23
sample_std_dev = 7

alpha = 0.05

# Z
z_score = (sample_mean - pop_mean)/(pop_std_dev/np.sqrt(sample_size))

p_value = 1-stats.norm.cdf(z_score)


# Z(alpha)
z_critical = stats.norm.ppf(1-alpha)

print(f"Z Score : {z_score}")
print(f"Z Critical : {z_critical}")


if z_score > z_critical:
  print("Decision : Accept H1 or Reject H0")
else:
  print("Decision : Failed to reject H0")


if p_value < alpha:
  print("Decision based on p-value : Accept H1 or Reject H0")
else :
  print("Decision based on p-value : Failed to reject H0")
```

Based on the given data and right tailed test, we can say that the average bill amount has increased significantly.
