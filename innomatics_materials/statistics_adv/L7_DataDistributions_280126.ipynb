## Bernouli Dist

```python
from scipy.stats import bernoulli
```

```python
## We have a data with category A(success) - 60 examples and category B(failure) - 40 examples
## P = 60/100 = 0.6
p=0.6

pmf_success = bernoulli.pmf(1,p) ## pmf(x=1,success)
pmf_failure = bernoulli.pmf(0,p)

print(pmf_success)
print(pmf_failure)
```

```python
## Lets say a probability of success is given to us, we can generate a random sample data using it

p=0.5

sample_data = bernoulli.rvs(p,size=20) #rvs - random variable sampling
sample_data
```

## Binomial Dist

```python
from scipy.stats import binom
p=0.5
n=10

sample_data = binom.rvs(n,p,size = 50)
sample_data

## here we are executing the random experiment 50 times to generate the data

## Our random experiment is - tossing a coin 10 times
```

```python
## prob of exactly 4 successes in 10 trails (single experiment with 10 trials)
pmf = binom.pmf(4,n,p)
pmf
```

```python
# prob of getting 4 or less than 4 successes in 10 trails
# P(x<=4)
cdf = binom.cdf(4,n,p)
print(cdf)
```

```python
## probability percentile function
ppf = binom.ppf(0.95,n,p)
ppf
```

## Multinominal Dist

```python
from scipy.stats import multinomial




## lets say we have a categorical data with cat A - 60% , B - 30% , C - 10%

n=10. # we are doing 10 trials in this random experiment
p_vals = [0.6,0.3,0.1]

# whats prob of getting A - 5 times, B - 4 times, C - 1 time
pmf = multinomial.pmf([5,4,1],n,p_vals)
print(pmf)
```

## Poisson Dist

```python
from scipy.stats import poisson
```

```python
# In a shop, on an average 3 customers come every 1 hr

l = 3

## Generate sample/simulation data for next 10 hrs
sample_data = poisson.rvs(l,size=10)
sample_data
```

```python
# whats prob of getting 2 customers in next 1 hr
pmf = poisson.pmf(2,l)
print(pmf)
```

```python
# whats prob of getting less or equal to 2 customers in next 1 hr
cdf = poisson.cdf(2,l)
print(cdf)
```

```python
## 90th percentile
ppf = poisson.ppf(0.9,l)
print(ppf)
```

## Uniform Dist

```python
from scipy.stats import uniform
```

```python
## We can generate data distribution if we have the data (we don't know its type and parameter)
## But if we don't have the data and have the type of distribution and its parameters then also we can generate data distribution
```

```python
loc,scale = 1,3

# pdf of uniform = 1/(b-a)

# loc = minimum value = a = 1
# scale = b-a = 3 (difference between max and min)
# for above dist b = 4

## we don't have data but we have its distribution type and its parameters
## using its dist type and parameter we can generate simulation data
```

```python
sample_data = uniform.rvs(loc=loc,scale=scale,size=10000)
sample_data
```

```python
import seaborn as sns
sns.histplot(sample_data,kde=True,bins=100)
```

```python
pdf = uniform.pdf(3.7,loc=loc,scale=scale)
pdf
```

```python
## 1/(b-a) = 1/3 = 0.33
```

```python
cdf = uniform.cdf(3.5,loc=loc,scale=scale)
cdf
```

```python
## ppf is inverse of cdf in case of continuous distributions
ppf = uniform.ppf(0.833,loc=loc,scale=scale)
ppf
```

```python
## if cdf(3.5) = 0.833
## then ppf(0.833) = 3.5
```

```python
uniform.cdf(2.5,loc=loc,scale=scale)
```

```python
uniform.ppf(0.5,loc=loc,scale=scale)
```

## Normal Dist

```python
from scipy.stats import norm
```

```python
mean,std_dev = 0,1
```

```python
sample_data = norm.rvs(loc=mean,scale=std_dev,size=10000)
sns.histplot(sample_data,kde=True,bins=50)
```

```python
pdf = norm.pdf(0.5,loc=mean,scale=std_dev)
pdf
```

```python
cdf = norm.cdf(0,loc=mean,scale=std_dev)
cdf
```

```python
## percent point function / prob percent point function

ppf = norm.ppf(0.97,loc=mean,scale=std_dev)
ppf
```
