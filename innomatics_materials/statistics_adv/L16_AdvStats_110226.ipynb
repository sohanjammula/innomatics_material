```python
'''
Problem Statement : Restaurant owner claims that 50% customers are male and 50% are female. we need to validate his claim.
We need to validate this claim with significance level of 0.05


H1 : proportion != 0.5

H0 : proportion = 0.5

'''
```

```python
from seaborn import load_dataset
tips = load_dataset('tips')
tips
```

```python
sample_size = 50
sample = tips.sample(n=sample_size)
sample
```

```python
observed_frequencies = sample['sex'].value_counts().values.tolist()
observed_frequencies
# from sample
```

```python
expected_frequencies = [sample_size*0.5, sample_size*0.5]
expected_frequencies
# claimed
```

```python
from scipy import stats
chi_sq_stat,p_value = stats.chisquare(f_obs = observed_frequencies, f_exp = expected_frequencies)
```

```python
chi_sq_stat,p_value
```

```python
alpha = 0.05
d_freedom = len(observed_frequencies)-1 #Degrees of Freedom
chi_square_critical = stats.chi2.ppf(1-alpha,d_freedom)
chi_square_critical
```

```python
if chi_sq_stat > chi_square_critical:
  print('Decision : Reject the H0')
else:
  print("Decision : Failed to reject the H0")
```

```python
'''
As we have failed to reject the H0 (in simple we can say we need to accept H0)
ie the claim made by owner that 50% customer are male and 50% are female is statisticaly significant
'''
```

## Kolmogorov-Smirnov Test
- Purpose : Test if the sample matches a given distribution
- When to use : Use for continous data to check goodness of fit for given continuous distribution
- The sample should be random and claimed distribution should be continuous


Example Claim : The tips['total_bill'] follows normal distribution.

```python
'''
Null Hypothesis H0 : the sample follows normal distribution
Alternate Hypothesis H1 : sample doesn't follow normal distribution

significance level = 0.05
'''
```

```python
alpha =0.05
sample = tips['total_bill']
ks_stat,p_value = stats.kstest(sample,'norm',args = (sample.mean(),sample.std()))

print(f"KS Stat={ks_stat}, p_value={p_value}")
```

```python
def decision(p_value,alpha):
  if p_value<alpha:
    print("Decision : Reject H0")
  else:
    print("Decision : Failed to reject the H0")
```

```python
decision(p_value,alpha)
# The data doesn't follow normal distribution
```

## One Sample t-test
- Purpose : Check if the claimed population mean is same as that of sample mean using a single sample
- When to use : Used for continous data, and comparing sample mean to a specific population mean
- Assumptions : Sample should be random, data follows normal distribution


Example claim : The average of 'tip' is 2.

```python
'''
H0 : The average tip is 2
H1 : The average tip is not 2
'''
```

```python
sample = tips['tip']
sample
```

```python
t_stat,p_value = stats.ttest_1samp(sample,2)

print(f"{t_stat=}, {p_value=}")
```

```python
decision(p_value,alpha)
```

## Pearson Correlation Test
- Purpose : Tests whether two continuous data have linear relationship or not
- When to use : Used for continous data to test linear relationship
- Assumptions: Data should be continous, normaly distributed

Example claim : 'total_bill' and 'tip' have linear relationship

```python
'''
H0 : The relationship is not linear

H1 : The relationship is linear
'''
```

```python
stat,p_value = stats.pearsonr(tips['total_bill'],tips['tip'])

print(f"{stat=} , {p_value=}")
```

```python
decision(p_value,alpha)
# As we reject H0, the relationship is significantly linear
```

## Two Sample t-test (Test of independence)

- Purpose : Compare means of two sample
- Use : Use for categorical and continous variable

Example Claim : The average of tip from Male and Female customers is same

```python
'''
H0 : The average tip from Male and Female customers are different
H1 : The average tip from Male and Female customers are same
'''
```

```python
sample_male = tips[tips['sex']=='Male']['tip']
sample_female = tips[tips['sex']=='Female']['tip']

stat,p_value = stats.ttest_ind(sample_male,sample_female)

print(f"{stat=} , {p_value=}")
```

```python
decision(p_value,alpha)

# The average tip from Male and Female customers are different
```

## One - way ANOVA
- Purpose : Compares means of three or more sample
- Use : Used for categorical data vs numerical data with >2 groups

Example Claim : Average of total_bill differs by day of the week.

```python
tips
```

```python
'''
H0 : The average is same for all days of week
H1 : The average differs for days of week
'''
```

```python
days = tips['day'].unique()
days

groups = [tips[tips['day']==day]['total_bill'] for day in days]
```

```python
stat,p_value = stats.f_oneway(*groups)
print(f"{stat=} , {p_value=}")
```

```python
decision(p_value,alpha)

# as we reject H0, the average differs significantly for days of the week
```
