# Polymorphism and Encapsulation
Python, being an object-oriented programming language, supports important OOP concepts like Polymorphism and Encapsulation. These concepts help in writing efficient and reusable code.

## 1. Polymorphism in Python

Polymorphism allows objects of different classes to be treated as objects of a common super class. It enables a single interface to represent different data types.

### Types of Polymorphism
`Method Overriding` (Runtime Polymorphism)

`Method Overloading` (Compile-time Polymorphism, which Python handles differently using default arguments and *args, **kwargs)

`Operator Overloading`

#### 1.1 Method Overriding

This occurs when a derived class provides a specific implementation of a method that is already defined in its parent class.

```python
class Animal:
    def sound(self):
        return "Some generic sound"
```

```python
class Dog(Animal):
    def sound(self):
        return "Bark"
```

```python
class Cat(Animal):
    def sound(self):
        return "Meow"
```

```python
animals = [Dog(), Cat(), Animal()]
```

```python
for animal in animals:
    print(animal.sound())  # Calls respective class's sound() method
```

#### Method Overloading (Using Default Arguments)

Python does not support traditional method overloading but achieves similar functionality using default values.

```python
class MathOperations:
    def add(self, a, b=0, c=0):
        return a + b + c
```

```python
math_obj = MathOperations()
print(math_obj.add(5))         # 5
print(math_obj.add(5, 10))     # 15
print(math_obj.add(5, 10, 15)) # 30
```

#### Operator Overloading

Python allows overriding the behavior of built-in operators using special method

```python
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y
    
    def __sub__(self, other):
        return Point(self.x * other.x, self.y * other.y)
    
    def __str__(self):
        return f"({self.x}, {self.y})"
```

```python
p1 = Point(2, 3)
p2 = Point(4, 5)
print(p1 - p2)  # Calls __add__ method and prints (6, 8)
```

```python
class point:
    def __init__(cl,x,y):
        cl.x=x
        cl.y=y
    def __mul__(cl,oth):
        return point (cl.x-oth.x,cl.y-oth.y)
    def __str__(cl):
        return f"{cl.x, cl.y}"
```

```python
__sub__ -
__mul__ *
__eq__ =
__add__ +
```

```python
point1 = point(10,8)
point2 = point(4,3)
print(point1 * point2)
```

# Encapsulation in Python

Encapsulation refers to restricting direct access to class data and methods to prevent accidental modification. It is achieved using:

Public Members (Accessible from anywhere)

Protected Members (_Single underscore: _var, Accessible within the class and subclasses)

Private Members (__Double underscore: __var, Accessible only inside the class)

```python
class Car:
    def __init__(self, brand, model, price):
        self.brand = brand      # Public
        self._model = model     # Protected
        self.__price = price    # Private
    
    def get_price(self):
        return self.__price  # Private variable accessed through method

    def set_price(self, price):
        if price > 0:
            self.__price = price
            
class car2(Car):
    def display(self):
        return self._model
    def get_price(self):
        return self.__price
    def get_model(self):
        return self._model
```

```python
car = Car("Tesla", "Model X", 80000)
```

```python
carr=car2("Tesla", "Model X", 80000)
```

```python
print(car._model)       # Accessible
print(carr.get_model())    # Accessible (not recommended to access directly)
```

```python
carr.display()
```

```python
#print(carr.__price)   # Raises an AttributeError
print(carr.get_price()) # Correct way to access private members
```

### Private Variables and Getter Methods
Problem:
Create a class BankAccount with a private variable __balance and a method get_balance().

```python
class bank:
    def __init__(cl,account,balance):
        cl.account=account
        cl.__balance=balance
    def deposit(cl,money):
        cl.__balance+=money
        return f"Account:{cl.account} deposited{money}"
    def withdraw(cl,money):
        cl.__balance-=money
        return f"Account:{cl.account} withdrawn{money}"
    def get_balance(cl):
        return f"balance available:{cl.__balance}"
```

```python
acc1=bank('1234',34000)
```

```python
print(acc1.__balance)
```

```python
acc1.deposit(3000)
```

```python
acc1.withdraw(1000)
```

```python
acc1.get_balance()
```

```python
class bankaccount:
    def __init__(cl,name,balance):
        cl.name=name
        cl.balance=balance
        cl._surchrg=0.02
        cl.__penalty=0.01
    def show_balance(cl):
        return (cl.balance-(cl.balance*cl._surchrg)-(cl.balance*cl.__penalty))
```

```python
cust1=bankaccount('ram',5000)
```

```python
cust1.show_balance()
```
