# Tasks of class object

```python
class rectangle:
    def __init__(cl,length,width):
        cl.length=length
        cl.width=width
    def area(cl):
        return cl.length*cl.width
    def perimeter(cl):
        return 2*(cl.length+cl.width)
```

```python
rect1=rectangle(3,4)
```

```python
print(rect1.area())
```

```python
print(rect1.perimeter())
```

```python
class library:
    def __init__(cl):
        cl.books=[] 
    
    def add_book(cl,title):
        cl.books.append(title)
        return f"you have added {title} book"
    
    def remove_book(cl,title):
        if title in cl.books:
            cl.books.remove(title)
            return f"book {title} has removed"
        return "book not found"
    def display(cl):
        return cl.books
            
```

```python
lib1=library()
```

```python
lib1.add_book('python_basics')
```

```python
lib1.add_book('Data science')
```

```python
lib1.display()
```

```python
lib1.remove_book('1984')
```

```python
lib1.remove_book('python_basics')
```

```python
lib1.display()
```

# Inheritance and Abstraction

## 1. Introduction to Inheritance
Inheritance is a fundamental concept in object-oriented programming (OOP) that allows a class (child class) to inherit attributes and methods from another class (parent class). It promotes code reusability and hierarchical relationships.

### Advantages of Inheritance:

Code Reusability: Avoids redundant code.

Extensibility: Child classes can extend functionalities of parent classes.

Data Organization: Establishes a hierarchical structure.

### 1.1 Types of Inheritance in Python
`Single Inheritance:` A child class inherits from a single parent class.

`Multiple Inheritance:` A child class inherits from multiple parent classes.

`Multilevel Inheritance:` A child class inherits from another child class.

`Hierarchical Inheritance:` Multiple child classes inherit from a single parent class.

`Hybrid Inheritance:` A combination of two or more types of inheritance.

#### 1.2 Single Inheritance Example

```python
class Parent:
    def show(self):
        print("This is the parent class.")
    @staticmethod
    def square(num):
        return num * num
```

```python
class Child(Parent):
    def display(self):
        print("This is the child class")
        
```

```python
# Creating an object of Child class
c = Child()
```

```python
c.square(3)  # Inherited from Parent class
```

```python
c.display()
```

#### 1.3 Multiple Inheritance Example

```python
class Mother:
    def mother_info(self):
        print("This is Mother class.")
```

```python
class Father:
    def father_info(self):
        print("This is Father class.")
```

```python

class Child(Father, Mother):
    def child_info(self):
        print("This is Child class.")
```

```python
c = Child()
```

```python
c.father_info()
```

```python
c.mother_info()
```

```python
c.child_info()
```

#### 1.4 Multilevel Inheritance Example

```python
class Grandparent:
    def grandparent_info(self):
        print("This is Grandparent class.")
```

```python
class Parent(Grandparent):
    def parent_info(self):
        print("This is Parent class.")
```

```python

class Child(Parent):
    def child_info(self):
        print("This is Child class.")
```

```python
c = Child()
```

```python
c.grandparent_info()
```

```python
c.parent_info()
```

```python
c.child_info()
```

#### 1.5 Hierarchical Inheritance Example

```python
class Parent:
    def show(self):
        print("This is the Parent class.")
```

```python
class Child1(Parent):
    def display1(self):
        print("This is Child1 class.")
```

```python
class Child2(Parent):
    def display2(self):
        print("This is Child2 class.")
```

```python
c1 = Child1()
```

```python
c1.show()
```

```python
c1.display1()
```

```python
c2 = Child2()
```

```python
c2.show()
```

```python
c2.display2()
```
