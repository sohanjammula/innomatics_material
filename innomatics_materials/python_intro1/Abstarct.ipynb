## Inheritance Examples

```python
class employee:
    def __init__(cl,name,salary):
        cl.name=name
        cl.salary=salary
    def show_details(cl):
        return f"employee name is {cl.name} and salary is {cl.salary}"
        
class manager(employee):
    def __init__(cl,name,salary,department):
        super().__init__(name,salary)
        cl.department=department
    def show_mgr_details(cl):
        return f"Manager name is {cl.name},salary is {cl.salary} in dpartment {cl.department}"
    
    
```

```python
mgr1=manager('ram',50000,'datascience')
```

```python
mgr1.show_details()
```

```python
mgr1.show_mgr_details()
```

```python
emp1=employee('ram',50000)
```

```python
emp1.show_details()
```

```python
class car:
    def __init__(cl,brand):
        cl.brand=brand
    def car_details(cl):
        return f" Car brand is {cl.brand}"
class electric_car(car):
    def __init__(cl,brand,battery):
        super().__init__(brand)
        cl.battery=battery
    def el_car_details(cl):
        return f"the car is {cl.brand} and the battery is{cl.battery}"
```

```python
car1=electric_car('tata','50000mah')
```

```python
car1.car_details()
```

```python
car1.el_car_details()
```

# Introduction to Abstraction

Abstraction is an OOP principle that hides the implementation details and only exposes the essential features. It is achieved in Python using abstract classes and methods via the abc module.

###  Abstract Classes and Methods

An abstract class is a class that contains one or more abstract methods.

An abstract method is a method that has only a declaration but no implementation in the parent class.

### Advantages of Abstraction:

Hides Complexity: Users only interact with essential functionalities.

Promotes Design Flexibility: Encourages modular design.

Encapsulation: Helps in achieving better security.

```python
from abc import ABC,abstractmethod
```

```python
class car(ABC):
    @abstractmethod
    def engine(cl):
        pass
    
```

```python
class automatic_car(car):
    def __init__(cl,name,milage):
        cl.name=name
        cl.milage=milage

    def cardetails(cl):
        return f"car is {cl.name}, and milage is {cl.milage}"
```

```python
car1=automatic_car('tata',25)
```

```python
class automatic_car(car):
    def __init__(cl,name,milage,engcap):
        cl.name=name
        cl.milage=milage
        cl.engcap=engcap

    def cardetails(cl):
        return f"car is {cl.name}, and milage is {cl.milage}"
        
    def engine(cl):
        return f"this car engine capacity is {cl.engcap}"
```

```python
car1=automatic_car('tata',25,'2000CC')
```

```python
car1.cardetails()
```

```python
car1.engine()
```

```python
class shape(ABC):
    @abstractmethod
    def area(cl):
        pass
```

```python
import math
class circle(shape):
    def __init__(cl,rad):
        cl.rad=rad
    def perimeter(cl):
        return 2*math.pi*rad
```

```python
cir1=circle(2)
```

```python
class circle(shape):
    def __init__(cl,rad):
        cl.rad=rad
    def perimeter(cl):
        return 2*math.pi*cl.rad

    def area(cl):
        return math.pi*cl.rad*cl.rad
```

```python
cir1=circle(2)
```

```python
cir1.perimeter()
```

```python
class Animal(ABC):
    @abstractmethod
    def food(cl):
        pass
class dog(Animal):
    def __init__(cl,breed,foods):
        cl.breed=breed
        cl.foods=foods
    def show_breed(cl):
        return f"this dog is {cl.breed}"
    def food(cl):
        return f"this dog eats {cl.foods}"
```

```python
class cow(Animal):
    def __init__(cl,foods):
        cl.foods=foods
    def food(cl):
        return f"this cow eats {cl.foods}"
```

```python
dog1=dog('lab','veg')
```

```python
dog1.food()
```

```python
dog1.show_breed()
```

```python
cow1=cow('grass')
```

```python
cow1.food()
```
