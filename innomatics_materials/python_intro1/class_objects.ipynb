# Classes and Objects
Python is an object-oriented programming language that allows defining and working with classes and objects. A class is a blueprint for creating objects, and an object is an instance of a class.

### Creating a Simple Class and Object

A class is defined using the class keyword, and an object is created by calling the class.

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def greet(self):
        return f"Hello, my name is {self.name} and I am {self.age} years old."
```

```python
person1 = Person("Alice", 25)
print(person1.greet())
```

### Class with Methods

A class can have methods that define the behavior of its objects.

```python
class Calculator:
    def __init__(z,a,b):
        z.a=a
        z.b=b

    def add(z):
        return z.a +z.b
    
    def subtract(z):
        return z.a - z.b
```

```python
calc = Calculator(3,4)
print(calc.add())      
# Output: 8
```

```python
print(calc.a)
```

```python
calc.subtract()
```

```python
calc = Calculator(5,3)
```

```python
print(calc.add())
```

```python
print(calc.subtract()) # Output: 6
```

### Class with Instance and Class Variables

Instance variables belong to individual objects.

Class variables are shared among all instances.

```python
class Employee:
    company = "TechCorp"  # Class variable
    
    def __init__(self, name, salary):
        self.name = name       # Instance variable
        self.salary = salary   # Instance variable
```

```python
emp1 = Employee("RAM", 50000)
emp2 = Employee("Jane", 60000)
```

```python
print(emp1.company)  # Output: TechCorp
print(emp2.company)  # Output: TechCorp
```

```python
print(emp1.name, emp1.salary) 
print(emp2.name, emp2.salary) 
```

### Class with Static Methods

A static method does not require an instance and is defined using @staticmethod.

```python
class Math:
    @staticmethod
    def square(num):
        return num * num
```

```python
cc=Math()
```

```python
print(cc.square(5))  # Output: 25
```

### Data Hiding with Private Attributes

Python allows defining private attributes using a double underscore __.

```python
class BankAccount:
    def __init__(self, balance):
        self.__balance = balance  # Private variable
    
    def get_balance(self):
        return self.__balance
```

```python
account = BankAccount(1000)
```

```python
print(account.__balance)  # Output: 1000
```

### Using Class as a Namespace

Classes can be used to group related methods together.

```python
class Config:
    DATABASE = "MySQL"
    PORT = 3306

print(Config.DATABASE)  # Output: MySQL
```

###  Create a Class for a Student
Problem:
Create a class Student with the following attributes and methods:

Attributes: name, roll_number, grade

Method: get_details() to return the student's details as a formatted string.

```python
class Student:
    def __init__(self, name, roll_number, grade):
        self.name = name
        self.roll_number = roll_number
        self.grade = grade

    def get_details(self):
        return f"Name: {self.name}, Roll Number: {self.roll_number}, Grade: {self.grade}"
```

```python
student1 = Student("Alice", 101, "A")
print(student1.get_details())
```
