```python
import numpy as np
import cv2 # library for computer vision to work with images
import matplotlib.pyplot as plt
```

## Read / open the image usinf cv2

```python
color_img = cv2.imread('/content/test_image.png')
color_img
```

```python
# cv2 reads image in BGR order but we need RGB for true colors
color_img = cv2.cvtColor(color_img,cv2.COLOR_BGR2RGB)
color_img
```

```python
color_img.shape
```

```python
bw_img = cv2.imread('/content/test_image.png',0)
# 0 is for loading image in gray scale
bw_img
```

```python
bw_img.shape
# bw image is a 2D array as it doesn't have 3 different color channels
```

## Draw on image

```python
line_img = cv2.line(color_img,(0,0),(200,200),(255,0,0),6)
# (0,0) - start of line coordinates
# (200,200) - end of line coordinates
# (255,0,0) - color of line
# 6 - thickness of line
line_img
```

```python
rect_img = cv2.rectangle(color_img,(250,100),(500,300),(255,0,0),6)
rect_img
```

## Saving image file

```python
cv2.imwrite('/content/rect_line_image.jpg',rect_img)
```

```python
processed_img = cv2.imread('/content/rect_line_image.jpg')
processed_img
```
